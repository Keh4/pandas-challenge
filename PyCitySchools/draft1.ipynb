```python
 # Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete_df.head()
```

```python
Total_Students = school_data_complete_df["student_name"].count()
Total_Students
```

```python
Total_Schools= school_data_df["school_name"].count()
Total_Schools
```

```python
Total_Budget= school_data_df["budget"].sum()
Total_Budget
```

```python
Average_Math_Score = school_data_complete_df["math_score"].mean()
Average_Math_Score
```

```python
Average_Reading_Score = school_data_complete_df["reading_score"].mean()
Average_Reading_Score
```

```python
# Create the bins in which Data will be held for Passing or Failing
# Bins are 0, 69.9, 100.   
bins = [0, 69.9, 100]

# Create the names for the five bins
group_names = ["fail", "pass"]
```

```python
school_data_complete_df["Math_P/F"] = pd.cut(school_data_complete_df["math_score"], bins, labels=group_names, include_lowest=True)
school_data_complete_df.head()
```

```python
school_data_complete_df["Reading_P/F"] = pd.cut(school_data_complete_df["reading_score"], bins, labels=group_names, include_lowest=True)
school_data_complete_df.head()
```

```python
percent_pass_math_df = school_data_complete_df.loc[school_data_complete_df["Math_P/F"]=="pass"]
percent_pass_math_df.head()
```

```python
 # Calculate the percentage of students who passed math
Pass_Math_count = len(percent_pass_math_df["Math_P/F"] == "pass")
Pass_Math_count
```

```python
Percent_Pass_Math_count = Pass_Math_count/Total_Students *100
Percent_Pass_Math_count
```

```python
percent_pass_reading_df = school_data_complete_df.loc[school_data_complete_df["Reading_P/F"]=="pass"]
percent_pass_reading_df.head()
```

```python
Pass_Reading_count = len(percent_pass_reading_df["Reading_P/F"] == "pass")
Pass_Reading_count
Percent_Pass_Reading_count = Pass_Reading_count/Total_Students *100
Percent_Pass_Reading_count
```

```python
Pass_both_count_df = percent_pass_reading_df.loc[(percent_pass_reading_df["Math_P/F"]=="pass") & 
                                                 (percent_pass_reading_df["Reading_P/F"]=="pass")] 
Pass_both_count_df
```

```python
Pass_both_count_total = Pass_both_count_df["student_name"].count()
Pass_both_count_total
```

```python
Pass_both_count_Percent = Pass_both_count_total/Total_Students*100
Pass_both_count_Percent
```
