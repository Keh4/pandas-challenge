```python
 # Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
#Set the Values for Summary - Totals Student, School and Budget, Avg Math and Reading Scores
Total_Students = school_data_complete_df["student_name"].count()
Total_Schools= school_data_df["school_name"].count()
Total_Budget= school_data_df["budget"].sum()
Average_Math_Score = school_data_complete_df["math_score"].mean()
Average_Reading_Score = school_data_complete_df["reading_score"].mean()
```

```python
# Create the bins in which Data will be held for Passing or Failing
# Bins are 0, 69.9, 100.   
bins = [0, 69.9, 100]

# Create the names for the bins
group_names = ["fail", "pass"]
```

```python
#Set df for pass/ fail
school_data_complete_df["Math_P/F"] = pd.cut(school_data_complete_df["math_score"], bins, labels=group_names, include_lowest=True)
school_data_complete_df["Reading_P/F"] = pd.cut(school_data_complete_df["reading_score"], bins, labels=group_names, include_lowest=True)
```

```python
#Set Values for % pass Reading and Math
percent_pass_math_df = school_data_complete_df.loc[school_data_complete_df["Math_P/F"]=="pass"]
Pass_Math_count = len(percent_pass_math_df["Math_P/F"] == "pass")
Percent_Pass_Math_count = Pass_Math_count/Total_Students *100
percent_pass_reading_df = school_data_complete_df.loc[school_data_complete_df["Reading_P/F"]=="pass"]
Pass_Reading_count = len(percent_pass_reading_df["Reading_P/F"] == "pass")
Percent_Pass_Reading_count = Pass_Reading_count/Total_Students *100
```

```python
#Set Values for % pass Overall
Pass_both_count_df = percent_pass_reading_df.loc[(percent_pass_reading_df["Math_P/F"]=="pass") & 
                                                 (percent_pass_reading_df["Reading_P/F"]=="pass")] 
Pass_both_count_total = Pass_both_count_df["student_name"].count()
Pass_both_count_Percent = Pass_both_count_total/Total_Students*100
```

```python
#District Summary
District_Summary_df = pd.DataFrame({"Total Schools":[Total_Schools], 
                                    "Total Students":[Total_Students], 
                                    "Total Budget":[Total_Budget],
                                   "Average Math Score":[Average_Math_Score],
                                   "Average Reading Score":[Average_Reading_Score],
                                   "% Passing Math":[Percent_Pass_Math_count],
                                   "% Passing Reading":[Percent_Pass_Reading_count],
                                   "% Overall Passing":[Pass_both_count_Percent]})


District_Summary_df
```

```python
#School Summary beginning df's to set values
grouped_school_df = school_data_complete_df.groupby(["school_name"])
school_comparison_mean_df = grouped_school_df.mean()
school_comparison_sums_df = grouped_school_df.sum()
```

```python
#Set the Values for Summary - Totals Student, School and Budget, Avg Math and Reading Scores
Total_Students_by_School = school_comparison_mean_df["size"]
Total_Budget_by_School = school_comparison_mean_df["budget"]
Per_Student_Budget = (Total_Budget_by_School)/ (Total_Students_by_School)
Avg_Math_Score_by_School = school_comparison_mean_df["math_score"]
Avg_Read_Score_by_School = school_comparison_mean_df["reading_score"]
```

```python
#Set Values for % pass Reading and Math
percent_pass_math_by_school_df = percent_pass_math_df.groupby(["school_name"])
count_pass_math_by_school_df = percent_pass_math_by_school_df.count()
count_pass_math_by_school = count_pass_math_by_school_df["Math_P/F"].astype('float')
percent_pass_math_by_school = (count_pass_math_by_school)/(Total_Students_by_School)*100
percent_pass_read_by_school_df = percent_pass_reading_df.groupby(["school_name"])
count_pass_read_by_school_df = percent_pass_read_by_school_df.count() 
count_pass_read_by_school = count_pass_read_by_school_df["Reading_P/F"].astype('float')
percent_pass_read_by_school = (count_pass_read_by_school)/(Total_Students_by_School)*100

```

```python
#Set Values for % pass Overall
percent_pass_both_by_school_df = Pass_both_count_df.groupby(["school_name"])
count_pass_both_by_school_df = percent_pass_both_by_school_df.count()
Pass_both_count_df.groupby("school_name").mean()
count_pass_both_by_school = count_pass_both_by_school_df["Math_P/F"].astype('float')
percent_pass_both_by_school = (count_pass_both_by_school)/(Total_Students_by_School)*100
school_by_type = Pass_both_count_df.drop_duplicates(["school_name"]).set_index(["school_name"])["type"]
```

```python
 #School Summary
School_Summary_df = pd.DataFrame({
                                    "School Type":school_by_type,
                                    "Total Students":Total_Students_by_School, 
                                    "Total School Budget":Total_Budget_by_School,
                                   "Per Student Budget":Per_Student_Budget,
                                   "Avg Math Score":Avg_Math_Score_by_School,
                                   "Avg Reading Score":Avg_Read_Score_by_School,
                                   "% Passing Math":percent_pass_math_by_school,
                                   "% Passing Reading":percent_pass_read_by_school,
                                   "% Overall Passing":percent_pass_both_by_school})  
School_Summary_df
```

```python
 #Top Performing Schools (By % Overall Passing)
    #Sort and display top 5 performing schools by overall passing
Top_Performing_School_Summary_df = School_Summary_df.sort_values("% Overall Passing", ascending=False)
Top_Performing_School_Summary_df.head(5)
```

```python
# Create the bins in which Data will be held for Spending Range per Student
  
bins = [0, 584.9, 629.9, 644.9, 680]

# Create the names for the bins
Spend_group_names = ["<$585", "$585-630", "$630-645","$645-680"]
```

```python
# Create the bins in which Data will be held for Scores by School size
  
bins = [0, 999.99, 1999.99, 5000]

# Create the names for the bins
School_Size_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```
